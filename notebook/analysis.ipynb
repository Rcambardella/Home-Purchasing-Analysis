```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = os.path.join('..', 'data', 'Analysis_of_homes_since_1984.xlsx')

dataset = pd.read_excel(file_path)

print(dataset.columns)

dataset.drop(['Unnamed: 4', 'Unnamed: 9'], axis=1, inplace=True, errors='ignore')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(data=dataset, x='YEAR', y=dataset['% OF PAYMENT COMPARED TO SALARY'] * 100)
plt.axhline(30, color='red', linestyle='--', label='Affordability Threshold (30%)')
plt.title('Mortgage Cost as % of Income (1984–2024)')
plt.ylabel('Mortgage Burden (%)')
plt.xlabel('Year')
plt.legend()
plt.grid(True)
plt.tight_layout()

chart_path = os.path.join('..', 'charts', 'housing_cost_burden.png')

plt.savefig(chart_path)

plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=dataset, x='Year', y='CPI ADJUSTED INCOME', label='CPI-Adjusted Income')
sns.lineplot(data=dataset, x='Year', y='Example of $97,600 Home', label='Home Price (Nominal)')
plt.title('Real Income vs Home Price (1984–2024)')
plt.ylabel('USD ($)')
plt.xlabel('Year')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
